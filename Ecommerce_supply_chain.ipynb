```python
#data cleaning
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
#getting year and month
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['month_name']=df['order_date'].dt.month_name()
```

```python
#demand_supply_gap
df['demand_supply_gap']= df['demand_units']-df['supply_units']
```

```python
#revenue,cost and profit

df['revenue']=df['demand_units']*df['selling_price']
df['cost']=df['supply_units']*df['unit_cost']
df['profit']=df['revenue']-df['cost']
```

```python
# Stock-out flag
df['stock_out_flag'] = np.where(df['inventory_level']==0,'Yes','No')
```

```python
df['demand_category']=pd.cut(df['demand_units'],bins=[0,100,250,500],labels=['low','medium','high'])
```
